```python
import pandas as pd
import numpy as np

# Simulated dataset (top 6 features + Target)
np.random.seed(42)
num_rows = 100

df = pd.DataFrame({
    "meaneduc": np.random.uniform(0, 15, num_rows),      # Average years of education
    "escolari": np.random.randint(0, 15, num_rows),      # Years of schooling
    "qmobilephone": np.random.randint(0, 3, num_rows),   # Number of mobile phones
    "rooms": np.random.randint(1, 7, num_rows),          # Number of rooms
    "r4m1": np.random.randint(0, 5, num_rows),           # Males under 12
    "overcrowding": np.random.uniform(0.1, 2.5, num_rows),  # Overcrowding index
    "Target": np.random.choice([1, 2, 3, 4], num_rows)    # Poverty level
})

df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

X = df.drop("Target", axis=1)
y = df["Target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

model = RandomForestClassifier()
model.fit(X_train, y_train)

print(classification_report(y_test, model.predict(X_test)))
```

```python
import gradio as gr

top_features = ["meaneduc", "escolari", "qmobilephone", "rooms", "r4m1", "overcrowding"]

def predict_poverty(*inputs):
    input_df = pd.DataFrame([inputs], columns=top_features)
    prediction = model.predict(input_df)[0]
    levels = {
        1: "üü• Extreme Poverty",
        2: "üüß Moderate Poverty",
        3: "üü® Vulnerable",
        4: "üü© Non-Vulnerable"
    }
    return f"Poverty Level: {levels[prediction]}"

gr.Interface(
    fn=predict_poverty,
    inputs=[gr.Number(label=feat) for feat in top_features],
    outputs="text",
    title="üè† Costa Rican Poverty Predictor",
    description="Simulated model predicting household poverty level using top features."
).launch()
```
